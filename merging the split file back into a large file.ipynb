```python
import pandas as pd
```

```python
#uploading split files

df1 = pd.read_csv('part1.csv')
df2 = pd.read_csv('part2.csv')
```

```python
#concatenating them vertically, row-wise
merged_df = pd.concat([df1,df2],ignore_index=True)   #ignore_index=True to reset or ignore the original row labels (index)

```

```python
#saving the restored merged file
merged_df.to_csv('merged_file.csv', index = False)  #index=False  tells pandas not to write the row index into the CSV.
```
