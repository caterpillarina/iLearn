```python
#Linear Programming
```

```python
#The problem:  
#Item      Wood    Labor(hrs)  Profit($)
#Table      3       2            50
#Chair      2       1            30

#We have 60 units of wood, 40 units of labor

#Goal is to Maximize profit
```

```python
#LP Formulation: 
#x= number of tables
#y= number of chairs
#Objective function: maximize Z =  50x+30y
#Constraints: 3x+2y <= 60 (wood)
#             2x+y <= 40(labor)
#             x,y >=  0
```

```python
from scipy.optimize import linprog

# Coefficients for the objective function (negative for maximization)
c = [-50, -30]

# Coefficients for the inequality constraints
A = [
    [3, 2],   # wood constraint
    [2, 1]    # labor constraint
]
b = [60, 40]  # resource limits

# Bounds for x and y (non-negative)
x_bounds = (0, None)
y_bounds = (0, None)

# Solve the LP
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Output
if result.success:
    print(f"Tables: {result.x[0]:.2f}, Chairs: {result.x[1]:.2f}")
    print(f"Maximum Profit: ${-result.fun:.2f}")
else:
    print("No solution found.")
```
