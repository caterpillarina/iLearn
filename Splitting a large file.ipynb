```python
#Github doesnot allow files larger than 25MB. I have a large file >42MB I wish to use in my code. 
#So breaking it into two smaller files. We are splitting row-wise
```

```python
import pandas as pd
df = pd.read_csv('downloads/observations.csv') #large file 
```

```python
midpoint = len(df) // 2
df1 = df.iloc[:midpoint]
df2 = df.iloc[midpoint:]

df1.to_csv('part1.csv',index = False)
df2.to_csv('part2.csv',index = False)
```
