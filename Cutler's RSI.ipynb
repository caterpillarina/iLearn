```python
import yfinance as yf
import pandas as pd

```

```python
symbol = "AApl"
df= yf.download(symbol, start="2020-01-01", auto_adjust =True) 
```

```python
# calculate returns. Day T2- DayT1. Close has to be Capital C. 
delta = df['Close'].diff() 
```

```python
gain = delta.where(delta>0, 0)
loss = -delta.where(delta<0,0)
```

```python
#Rsi period is set to 14
window = 14
```

```python
#rollign averages to smooth the gains and loss data. We are doing Cutler's RSI- meaning we are smoothing with Simple Moving Averages. (SMA) 

avg_gain = gain.rolling(window=window, min_periods=window).mean()
avg_loss = loss.rolling(window=window, min_periods=window).mean()
```

```python
#Finding rs
rs = avg_gain / avg_loss
```

```python
#finding rsi by normalizing rs 

df['RSI'] = 100 - (100 / (1 + rs))

print(df[['Close', 'RSI']].tail(20))
```
