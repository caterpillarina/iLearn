```python
#This code was created by a human with help from AI

import pandas as pd
```

```python
df = pd.read_csv('downloads/global_recipes.csv')
```

```python
#converting ingredients string into Python set. Not a list ? 

df['ingredients_set'] = df['Ingredients'].apply(lambda x: set(i.strip().lower() for i in x.split(',')))
                                                
```

```python
#practicing again 
#df['ingredients_set'] = df['Ingredients'].apply( lambda x : set(i.strip().lower() for i in x.split(',')))

#why a set and not a list ? 

#List allows duplicates. Set doesnt
#list -> ordered data, Sets -> Unique items , fast look up
#sets are faster for membership tests
#both are mutable

```

```python
#What ingredients are available ?
user_input = input('What are your available ingredients? Please type each ingredients separated by commas: ')


```

```python
#Convert the user input into a list of cleaned data. Strip header and trailer space with strip(), and split using split(',')

#  avail_ingredients = word.strip().lower() for word in user_input.split(',') 

#This gave an invalid syntax error because I was trying to use a generator expression or list comprehension without wrapping it in a container like list() or set(). 

#fixed it by adding square brackets to make it a list, or flower brackets fro a set.
#We are using sets. because we dont want duplciates. SO {}


```

```python
avail_ingredients = {word.strip().lower() for word in user_input.split(',') }
```

```python
# Creating a dataframe with List of recipes that can be cooked. #
#This is established by checking if a recipes ingredients are a SUBSET of my avail_ingredients
```

```python
df['can_cook']= df['ingredients_set'].apply(lambda x: x.issubset(avail_ingredients)) 

#what if i make my ingredients list  a subset of available ingredients . tryign in next row
```

```python
recipes_you_can_make = df[df['can_cook']][['Recipe Name', 'Country of Origin', 'Time to Cook']]
```

```python
#recipes you can almost make
df['missing'] = df['ingredients_set'].apply(lambda x : x - avail_ingredients)
```

```python
df['missing_count'] = df['missing'].apply(len)
```

```python
recipes_you_can_make_with_adjustments = df[df['missing_count'] == 1][['Recipe Name','Country of Origin', 'Time to Cook','missing']]
```
